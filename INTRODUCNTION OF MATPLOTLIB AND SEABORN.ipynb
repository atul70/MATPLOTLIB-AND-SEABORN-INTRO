## matplotlib And Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## matplotlib works on list , array , and pandas series....not work on pandas dataframe....

## When we use visualization for pandas Data Frame...then we Use to Seaborn

```python
# make a chart b/w lists
year =[2012,2013,2014,2015,2016]
sales=[23433,343544,44000,56565,66600]
```

```python
plt.plot(year,sales)
```

```python
customers=['ram','shyam','hari','prassad','sukhi']
profit=[23000,34000,45000,54443,43423]
```

```python
plt.plot(customers,sales,'green')
plt.plot(customers,profit,'red')
plt.xlabel('customers')
plt.ylabel('sales and profit')
plt.title('CUSTOMER VS SALES AND PROFIT')
```

```python
# when we make two charts of customer vs sales   and customers vs profit

plt.axes([0.05,0.05,0.9,0.9])
plt.plot(customers,sales,'red')
plt.xlabel('CUSTOMERS')
plt.ylabel('SALES')
plt.title('CUSTOMERS VS SALES')
plt.axes([1.2,0.05,0.9,0.9])
plt.plot(customers,profit,'green')
plt.xlabel('CUSTOMERS')
plt.ylabel('PROFITS')
plt.title('CUSTOMERS VS PROFIT')
```

```python
plt.subplot(2,1,1)
plt.plot(customers,sales,'red')
plt.xlabel('customers')
plt.ylabel('sales')
plt.title('Customers Vs Sales')
plt.subplot(2,1,2)
plt.plot(customers,profit,'green')
plt.xlabel('customers')
plt.ylabel('profit')
plt.title('Customers Vs Profits')
plt.tight_layout()
```

## Zoomed Version Of Data ..comapre between any two customers 

```python
plt.plot(customers,sales)
plt.xlabel('Customer')
plt.ylabel('sales')
plt.xlim('ram','shyam')
plt.title('Customers Vs Sales')
```

```python
# another way of zooming the chart is
plt.plot(customers,profit)
plt.xlabel('Customers')
plt.ylabel('Profit')
plt.title('Customer VS profit')
plt.axis(('ram','shyam',23000,30000))
```

## scatter plot

```python
customers=['ram','shyam','hari','prassad','sukhi']
profit=[23000,34000,45000,54443,43423]
sales=[23433,343544,44000,56565,66600]
plt.scatter(sales,profit,marker ='*',color='green',label='Sales Vs Profit')
plt.annotate('ram',xy=(23000,23433))   # this is mark on point,,where ram is existed                                         
plt.legend(loc=1)
plt.xlabel('Sales')
plt.ylabel('profits')
```

```python

# distribute histogram chart

fig,(ax0,ax1)=plt.subplots(nrows=1,ncols=2,figsize=(12,7))
sns.distplot(total['AirTime'],ax=ax0)
sns.distplot(total['AirTime'],ax=ax1)
ax1.set(xlabel='AirTime',xlim=(100,500))
ax1.axvline(x=170,label='MyTarget',linestyle='--')
ax1.legend()
```

## Now we plot 2 more chart 

```python
# this is linear model plot used in regression model ...it is regplot
# reg means regression line plot
# y=mx+c
# in this plot we cant use  ,, hue and col function...if we use hue and col function in regplot,,give error

sns.regplot(x='ActualElapsedTime',y='AirTime',data=total)
```

```python
# Now using another important plot,,this plot also show regression model...but in this plot we can add and change it colours
# and make plots according to category

sns.lmplot(x='ActualElapsedTime',y='CRSElapsedTime',data=total,hue='UniqueCarrier',col='UniqueCarrier')
```

```python
# now next plot is joinplot...its give scatter diagram with histogram

sns.jointplot(x='ActualElapsedTime',y='CRSElapsedTime',data=total)
```

```python
# now we make count plot..its find the count of entry of categorywise

sns.countplot(data=total,y='AirTime',hue='UniqueCarrier')
```

```python
# barplot 

sns.barplot(data=total,y='AirTime',x='UniqueCarrier')
```

```python
# strip plot...which shows density of any columns

sns.stripplot(y='AirTime',x= 'UniqueCarrier',data=total)
```

```python
# box and whisker plot

sns.boxplot(y='AirTime',x= 'UniqueCarrier',data=total)
```

```python
# now we create   pairplot

sns.pairplot(total)
```

```python
# corelation of pairplot chart

total.corr()
```

```python
# now we can make a chart of above ratio between variables...heatmap

sns.heatmap(total.corr(),annot=True)
```
